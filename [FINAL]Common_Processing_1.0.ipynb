# ## Basic Common Processing

#### Target variable & Useless Columns

```python
# Reassign target
d.Attrition.replace(to_replace = dict(Yes = 1, No = 0), inplace = True)

# Find the useless features
li_useless_feat = []
for col in list(d.columns):
    if d[col].nunique() == 1:
        li_useless_feat.append(col)
print('Useless Features:{}'.format(li_useless_feat))

# Delete the useless features
d.drop(columns=li_useless_feat, inplace=True)
d.head(2)
```

#### Generation-related Columns & Data Split
 - Generation-related Columns: [aaa, bbb, ccc] 
 - Data Split: [data_gen_x, data_gen_y, data_gen_z]

```python
def age_to_born_yyyy(x):
    return 2015 - x

def cat_generation(x):
    if (x>=1940) & (x<=1959):
        return 'gen_baby_boomer'
    elif (x>=1960) & (x<=1979):
        return 'gen_x'
    elif (x>=1980) & (x<=1994):
        return 'gen_y'
    elif (x>=1995) & (x<=2010):
        return 'gen_z'
    else:
        return 'gen_alpha'        
```

```python
# Add generation-related columns
d['born_yyyy'] = d.Age.apply(age_to_born_yyyy)
d['generation'] = d.born_yyyy.apply(cat_generation)
```

```python
display(
    np.round(d.generation.value_counts()/len(d)*100, 2),
    d.head(2)
)
```

```python
d.to_csv('./data/data_pped_2.0.csv', index=False, header=True)
```

```python
# separate the dataframe regarding generations
d_x = d[d.generation=='gen_x'].reset_index(drop=True)
d_y = d[d.generation=='gen_y'].reset_index(drop=True)
d_z = d[d.generation=='gen_z'].reset_index(drop=True)

# separate the data
d_x.to_csv('./data/gen_x_data.csv', index=False, header=True)
d_y.to_csv('./data/gen_y_data.csv', index=False, header=True)
d_z.to_csv('./data/gen_z_data.csv', index=False, header=True)
```
